```python
import pandas as pd

# File path
filepath = "Novi_Izveštaj o radu_za_Nastavnike_Natasa_Bogdanovic.xlsx"

# Sheet names to extract
sheet_names = ["oktobar 2024"]

# Read the Excel file
try:
    sheets = pd.read_excel(filepath, sheet_name=sheet_names, header=None)
except Exception as e:
    print(f"Error reading Excel file: {e}")
    exit()

# Dictionary to store extracted data
extracted_data = {}

# Predefined sections
sections = [
    "Kvalitet nastavnog procesa",
    "Rad sa Studentima",
    "Podizanje kvaliteta ustanove",
    "Jačanje kapaciteta i imidža ustanove",
    "Ostalo"
]
sections_lower = [s.lower() for s in sections]

# Process each sheet
for sheet, df in sheets.items():
    print(f"Processing sheet: {sheet}")

    # Convert all cells to strings and handle empty cells
    df = df.map(lambda x: str(x).strip() if pd.notna(x) and str(x).strip() != "" else "")

    # Initialize section data
    section_data = {section: [] for section in sections}
    current_section = None

    # Track the last occurrence of "Ostalo"
    last_ostalo_index = -1

    # Find the last occurrence of "Ostalo"
    for i, row in enumerate(df.values.tolist()):
        row_cells = [cell for cell in row if cell]  # Remove empty cells
        if not row_cells:
            continue

        # Check if the first cell is "Ostalo"
        first_cell = row_cells[0].lower()
        if first_cell == "ostalo":
            last_ostalo_index = i

    # Process data rows
    for i, row in enumerate(df.values.tolist()):
        row_cells = [cell for cell in row if cell]  # Remove empty cells
        if not row_cells:
            continue

        # Detect sections
        first_cell = row_cells[0].lower()
        if first_cell in sections_lower:
            # If "Ostalo" is the last occurrence, treat it as a section header
            if first_cell == "ostalo" and i == last_ostalo_index:
                current_section = sections[sections_lower.index(first_cell)]
            # If "Ostalo" appears earlier, treat it as regular content
            elif first_cell == "ostalo" and i != last_ostalo_index:
                section_data[current_section].append(" ".join(row_cells))
            # For all other section headers, treat them as section headers
            else:
                current_section = sections[sections_lower.index(first_cell)]
            continue  # Skip section header row

        # Add data to the current section
        if current_section:
            section_data[current_section].append(" ".join(row_cells))

    # Store the data in the dictionary
    extracted_data[sheet] = section_data

# Debugging: Print the extracted data structure
print("\nExtracted Data Structure:")
for sheet, content in extracted_data.items():
    print(f"Sheet: {sheet}")
    for section, text_list in content.items():
        print(f"Section: {section}")
        for text in text_list:
            print(f"  - {text}")
```

```python
import pandas as pd

def extract_text_data(file_path, sheet_name='Osnovni podaci'):
    # Load the Excel sheet
    df = pd.read_excel(file_path, sheet_name=sheet_name, dtype=str)  # Read all as string to preserve text
    
    # Drop completely empty columns
    df.dropna(axis=1, how='all', inplace=True)
    
    # Convert NaN values to empty strings
    df.fillna('', inplace=True)
    
    return df

# Example usage
file_path = "Novi_Izveštaj o radu_za_Nastavnike_Natasa_Bogdanovic.xlsx"  # Change this to your file path
data = extract_text_data(file_path)

# Print extracted data in the same structure as in Excel
print(data.to_string(index=False))

```
